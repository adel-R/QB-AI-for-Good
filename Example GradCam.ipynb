# Getting heatmap Mobilenet

```python
path_to_img = 'images/plume/20230111_methane_mixing_ratio_id_8714.tif'
path_to_models = "models/ensemble_witth_artifical_test/"
best_model = 0
model_name = f"best_mobilenet_v3_large_fold_{best_model}.pt"
device = GradCam.get_device()
```

```python
model = inference.load_mobilenet_v3_large(path_to_models+model_name)
model.to(device)
model.eval()
print("")
```

```python
heat_map, img_raw, lbl = GradCam.get_heatmap(model=model, path_to_img=path_to_img, device=device, against_label=None, model_type="MobileNet",layer_number=9)
fig = GradCam.visualize_heatmap(img_raw, heat_map, lbl)
```
